## Práctico 3: Frozen Lake

### Dependencias

```python
from gymnasium.envs.toy_text.frozen_lake import generate_random_map
from model import Model
from main import main
from frozen_lake_env import FrozenLakeEnv
```

### Probando el ambiente

Recomendamos comentar esta parte cuando ejecuten UCS y A*

```python
env = FrozenLakeEnv(
            "FrozenLake-v1",
            desc=generate_random_map(size=4), # genera un mapa aleatorio de dimension 4x4
            is_slippery=False, # probabilidad 0 de resbalar (100% determinista)
            render_mode="human", # muestra el entorno en pantalla ("ansi" para imprimir el entorno en la consola)
        )
```

Action Space y Observation Space

```python
print("Action Space: ", env.action_space)
print("Observation Space: ", env.observation_space)

print("Accion aleatoria: ", env.action_space.sample())
print("Observacion aleatoria: ", env.observation_space.sample())
```

Imprimir el laberinto

```python
print(env.map())
```

```python
obs = env.reset()
x_state, y_state = env.state_to_coords(obs)

print("El estado es: ", obs, " y sus coordenadas son: ", x_state, y_state)
```

Jugando aleatoriamente

```python
done = False
env.reset()
env.render()

while not done:
    action = env.action_space.sample()
    obs, reward, done = env.step(action)
    x_state, y_state = env.state_to_coords(obs)
    print(f"Accion: {action}, Observacion: {obs}, Recompensa: {reward}, Hecho: {done}")
    env.render()
```

```python
model = Model(env.map())
model.graph
```

Obtener las posiciones *x* e *y* de un estado

```python
obs = env.reset()
x_state, y_state = env.state_to_coords(obs)
print("Estado inicial: ", obs, " Coordenadas: ", x_state, y_state)
```

```python
initial_state = env.initial_state()
x_state, y_state = env.state_to_coords(initial_state)
print("Estado inicial: ", initial_state, " Coordenadas: ", x_state, y_state)
```

```python
goal = env.goal()
x_state, y_state = env.state_to_coords(goal)
print("Estado meta: ", goal, " Coordenadas: ", x_state, y_state)
```

```python
from search_agent import SearchAgent
search_agent = SearchAgent(env=env, initial_state=initial_state, end_state=goal, model=model)
```

```python
search_agent.run()
```
